```python
import json
file_name = '메타버스_naver_news'
with open("./"+file_name+'.json', encoding='utf8') as j_f:
    data=json.load(j_f)
print(data)
```

```python
data_description = []

for item in data:
    data_description.append(item['description'])
data_description
```

```python
import pandas as pd

data_df = pd.DataFrame({'description':data_description})
```

```python
import re

data_df['description'] = data_df['description'].apply(lambda x:re.sub(r'[^ ㄱ - | 가-힣]+', " ", x))
data_df.head()
```

```python
description = data_df['description']
```

```python
from konlpy.tag import Okt
okt = Okt()

description_noun = []

for d in description:
    description_noun.append(okt.nouns(d))
```

```python
print(description_noun)
```

```python
description_noun_join = []

for d in description_noun:
    d2 = [w for w in d if len(w) > 1]
    description_noun_join.append(" ".join(d2))
```

```python
print(description_noun_join)
```

```python
def okt_tokenizer(text):
    tokens = okt.morphs(text)
    return tokens

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(tokenizer = okt_tokenizer, min_df=2)
dtm = tfidf.fit_transform(description_noun_join)
```

```python
vocab = dict()

for idx, word in enumerate(tfidf.get_feature_names()):
    vocab[word] = dtm.getcol(idx).sum()
    
words = sorted(vocab.items(), key=lambda x:x[1], reverse=True)
```

```python
words
```

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

fm.get_fontconfig_fonts()
font_location = 'C:/Windows/Fonts/malgun.ttf'
font_name = fm.FontProperties(fname = font_location).get_name()
matplotlib.rc('font', family = font_name)

max = 30
```

```python
plt.figure(figsize = (12, 5))
plt.bar(range(max), [i[1] for i in words[:max]])
plt.xlabel("키워드", fontsize=12)
plt.ylabel("TF-IDF의 합", fontsize=12)
plt.grid(True)
plt.xticks(range(max), [i[0] for i in words[:max]], rotation=70)

plt.show()
```
